```python
import json 
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("stor_data.json")
```

```python
print(type(data))
print(data)
```

```python
def clean_data(data):
    text_to_num = {"one" : 1, "two": 2, "three":  3, "four":  4, "five":  5}
    cleaned_data = []
    unique_users = set()

    
    for user in data:
         raw_rating = str(user["rating"]).strip().lower()
         if(raw_rating in text_to_num):
             raw_rating = text_to_num[raw_rating]
        

         user["rating"] = raw_rating


        # handle missing value
         raw_age = user.get("age")
         if(raw_age == None):
            user["age"] = None

        #duplication
         if(user["name"].strip() in unique_users):
            continue

         unique_users.add(user["name"])
         cleaned_data.append(user)


    return cleaned_data
         
        
    
```

```python
data = clean_data(data)
print(data)
```

```python
def get_insights(data):

    total_rating = 0

    for user in data:
        total_rating += float(user["rating"])

    print(f"average rating of user is = {total_rating/(len(data))}")



    poor_rating = 0
    poor_ratings = 0


    for user in data:
        if(float(user["rating"]) < 3):
            poor_rating += float(user["rating"])
            poor_ratings += 1
            

    print(f"the percentage of poor rating users = {(poor_rating/len(data))*100}%")
    print(f"the percentage of poor rating = {(poor_ratings/len(data))*100}%")
        
```

```python
get_insights(data)
```
